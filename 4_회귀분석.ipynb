```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = np.random.randint(1,99,size=(10,2)).tolist()
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
x_data = np.array(x)
y_data = np.array(y)

a = 0
b = 0
lr = 0.0002
epochs = 2001

def mse(y_pred, y_data):
    return ((y_pred - y_data) ** 2).mean()

def mse_val(y_pred, y_data):
    return mse(y_pred, y_data)

for i in range(epochs):
    y_pred = a * x_data + b
    error = y_data - y_pred
    a_diff = -(1/len(x_data)) * sum(x_data * (error))
    b_diff = -(1/len(x_data)) * sum(error)
    a = a - lr * a_diff
    b = b - lr * b_diff

    mse_value = mse_val(y_pred, y_data)


    if i > 0 and mse_value > prev_mse:
        lr = max(lr - 0.00001, 0.00001)

    if i % 400 == 0:
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f, LR:%.04f, a_diff:%.04f, b_diff:%.04f, MSE=%.08f" % (i, a, b, lr, a_diff, b_diff, mse_value))

    prev_mse = mse_value
```
