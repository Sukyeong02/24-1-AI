```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

np.random.seed(3)
tf.random.set_seed(3)

Data_set = np.loadtxt('/content/drive/MyDrive/24-1학기 코딩 수업/인공지능/ThoraricSurgery.csv', delimiter=',')

X = Data_set[:,0:17]
Y = Data_set[:,17]

model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

test_X = X[:5, :]
test_Y = Y[:5]

predictions = model.predict(test_X)

for i in range(len(test_Y)):
    print('Actual Class: %.f, Predict Class: %.08f' % (test_Y[i], predictions[i][0]))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=100, batch_size=100)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

np.random.seed(3)
tf.random.set_seed(3)

Data_set = np.loadtxt('/content/drive/MyDrive/24-1학기 코딩 수업/인공지능/ThoraricSurgery.csv', delimiter=',')

X = Data_set[:,0:17]
Y = Data_set[:,17]

model = Sequential()
model.add(Dense(50, input_dim=17, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=1000, batch_size=10)
```
