### 실습과제 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/24-1학기 코딩 수업/인공지능/pima-indians-diabetes.csv", names = ["pregnant","plasma","pressure","thickness","insulin","BMI","pedigree","age","class"])
```

```python
data = df[['pregnant','plasma']]
print(type(data))
```

```python
data1 = data.values
print(data1)
```

```python
data2 = data1[:5]
print(data2)
print(data2.shape)
print(type(data2))
```

```python
x1 = []
for row in data2:
  x1.append(row[0])
print(x1)

y1 = []
for row in data2:
  y1.append(row[1])
print(y1)
```

```python
plt.plot(x1, y1, "o")
plt.plot(x1, y1, c='blue')
plt.show()
```

### 실습과제3

```python
Average_age = df['age'].mean()
print(Average_age)
```

```python
df2 = df[df['age'] <= Average_age]
df3 = df[df['age'] > Average_age]

df2.thickness.plot.hist(bins=50, alpha=1.0, color="skyblue")
df3.thickness.plot.hist(bins=50, alpha=0.2, color="orange")
plt.legend(['low_age', 'high_age'], loc='upper right')
plt.xlabel('thickness')
plt.show()
```
